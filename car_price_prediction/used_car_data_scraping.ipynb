```python
# Storing URL in a variable
url_c24=('https://www.cars24.com/')
```

```python
# Creating a function to extract data from Cars24.com
def cars24_list(url):
    
    # Establishing Connection with the website
    driver.get(url_c24)
    time.sleep(2)

    # Closing the location bar is appeared
    try:
        driver.find_element_by_xpath("//div[@class='float-right TBxAN']").click()
    except NoSuchElementException:
        pass
    
    # Clicking on Buy used car tab which is the 3rd option on the header
    driver.find_element_by_xpath("//ul[@class='float-right _1yq3e']/li[3]").click()
    time.sleep(2)
    
    # Creating a list to store urls
    urls=[]
    
    # Creating a for loop to iterate through cities to fetch the data
    for i in range(1,11):
        
        # Clicking on the location dropdown
        driver.find_element_by_xpath("//div[@class='col-md-3 _2VocW']/p/span").click() 
        time.sleep(2)
        
        # Clicking one city at a time
        driver.find_element_by_xpath(f"//ul[@class='_16Bvy']/li[{i}]").click()
        time.sleep(2)
        
        
        # Defining a function to iterate through bodytype
        for fltr in range(1,6):
            
            # Clicking More Filter button
            driver.find_element_by_xpath("//ul[@class='_1bWR_']/li[6]/div").click()
            time.sleep(1)
            
            # Selecting a bodytype
            driver.find_element_by_xpath(f"//div[@class='dJM1f']/div[3]/ul/li[{fltr}]").click()
            time.sleep(1)
            
            # Cliking SHOW CARS button
            driver.find_element_by_xpath("//button[@class='btn _3BmFU AgO6q']").click()
            time.sleep(2)
            
            # Scrolling to get more cars in view
            for x in range(0,12):
                driver.execute_script('window.scrollBy(0,1000)')
                time.sleep(1)
            
            # Collecting link for each bodytype for each city
            for link in driver.find_elements_by_xpath("//div[@class='col-4']/a"):
                urls.append(link.get_attribute('href'))
            
            # Clicking Clear All Filter button to clear filters  
            driver.find_element_by_xpath("//span[@class='_1QeRC']").click()
            time.sleep(1)
    
    # Creating Empty list to store data
    
    cars24_name=[]
    cars24_brand=[]
    cars24_variant=[]
    cars24_transmission=[]
    cars24_year=[]
    cars24_kms=[]
    cars24_fuel=[]
    cars24_owner=[]
    cars24_location=[]
    cars24_price=[]
    
    # Collecting data for each url
    
    for url in urls:
        driver.get(url)
        time.sleep(3)

        try:

            cars24_year.append(driver.find_element_by_xpath("//div[@class='_10kZS']/h2").text.split()[0])

        except NoSuchElementException:

            cars24_year.append('-')

        try:
            cars24_brand.append(driver.find_element_by_xpath("//div[@class='_10kZS']/h2").text.split()[1])

        except NoSuchElementException:
            
            cars24_brand.append('-')

        try:
            name = driver.find_element_by_xpath("//div[@class='_10kZS']/h2").text.split()[2:4]
            cars24_name.append(' '.join(name))

        except NoSuchElementException:
            
            cars24_name.append('-')

        try:
            cars24_variant.append(driver.find_element_by_xpath("//div[@class='_10kZS']/h2/span").text.split()[0])

        except NoSuchElementException:
            
            cars24_variant.append('-')

        try:
            cars24_transmission.append(driver.find_element_by_xpath("//div[@class='_10kZS']/h2/span/small").text)            

        except NoSuchElementException:
            
            cars24_transmission.append('-')

        try:
            cars24_kms.append(driver.find_element_by_xpath("//ul[@class='_2tp8h']/li[1]").text.replace('km','').replace(',',''))

        except NoSuchElementException:
            
            cars24_kms.append('-')

        try:
            cars24_fuel.append(driver.find_element_by_xpath("//ul[@class='_2tp8h']/li[3]").text)

        except NoSuchElementException:
            
            cars24_fuel.append('-')

        try:
            cars24_owner.append(driver.find_element_by_xpath("//ul[@class='_2tp8h']/li[5]").text)

        except NoSuchElementException:
            
            cars24_owner.append('-')

        try:
            cars24_location.append(driver.find_element_by_xpath("//p[@class='_2NHUv']/span").text)

        except NoSuchElementException:
            
            cars24_location.append('-')

        try:
            cars24_price.append(driver.find_element_by_xpath("//div[@class='text-right g-hPT']/strong").text.replace('â‚¹','').replace(',',''))

        except NoSuchElementException:
            
            cars24_price.append('-')
```
