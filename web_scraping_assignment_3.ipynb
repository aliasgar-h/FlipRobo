```python
# Defining function to scrape data according to product, pincode and number of pages

def amazon_buy(product,pincode,nop):
    # creating a URL variable to store website URL
    url_amzn=('https://www.amazon.in/')
    
    # Establishing Connection with the website
    driver.get(url_amzn)
    time.sleep(2) # pausing the code execution for 2 seconds for website to load

    # Creating object of the search bar by finding it through ID
    srch_bar=driver.find_element_by_id('twotabsearchtextbox')

    # Clearing the bar if there is any text in it
    srch_bar.clear()

    # Sending required product to the search bar
    srch_bar.send_keys(product)

    # Creating onject of the search button and sending click commamnd
    srch_btn=driver.find_element_by_id('nav-search-submit-button')
    srch_btn.click()
    time.sleep(2)
    
    
    # Finding and clicking on the pincode item
    driver.find_element_by_id('glow-ingress-line2').click()
    time.sleep(1)
    
    # Finding the search bar to enter the desired city pincode
    pin_search=driver.find_element_by_id('GLUXZipUpdateInput')
    
    # Clearing the bar
    pin_search.clear()
    # Sending user provided pincode to the bar
    pin_search.send_keys(pincode)
    time.sleep(1)
    # Clicking the apply button
    driver.find_element_by_id('GLUXZipUpdate').click()
    time.sleep(2)
    
    # Collecting URLs for each product on the page
    
    for i in range (0,nop):
        for i in driver.find_elements_by_xpath("//h2[@class='a-size-mini a-spacing-none a-color-base s-line-clamp-2']/a|//h2[@class='a-size-mini a-spacing-none a-color-base s-line-clamp-4']/a"):
            links.append(i.get_attribute('href'))
        driver.find_element_by_xpath("//li[@class='a-last']/a").click()
        time.sleep(2)
        
    # Iterating through the Links list and accessing each link to extract data
    for link in links:
        driver.get(link)
        time.sleep(2)
    
        # Extractibng product brand
        try:
            product_brand.append(driver.find_element_by_id('bylineInfo').text)
        except NoSuchElementException:
            product_brand.append('-')

        # Extracting Product Title
        product_name.append(driver.find_element_by_xpath("//span[@class='a-size-large product-title-word-break']").text)

        # Extracting Product Rating
        try:
            product_rating.append(driver.find_element_by_xpath("//span[@class='a-size-base a-nowrap']/span").text)

        except NoSuchElementException:
            product_rating.append('-')

        # Extracting the total number of ratings for the product
        try:
            no_of_rating.append(driver.find_element_by_id('acrCustomerReviewText').text)
        except NoSuchElementException:
            no_of_rating.append('-')

        # Extracting Product Price. There are three different prices DealPrice, Amazon'sPrice, SalePrice. We search for anyone of the following. If none is found we append No Price available.
        try:
            product_price.append(driver.find_element_by_xpath("//span[@class='a-size-medium a-color-price priceBlockDealPriceString']|//span[@class='a-size-medium a-color-price priceBlockBuyingPriceString']|//span[@class='a-size-medium a-color-price priceBlockSalePriceString']").text)

        except NoSuchElementException:        
            product_price.append('-')

        # Extracting Return/Exchange policy data
        try:
            ret_exch.append(driver.find_element_by_id('RETURNS_POLICY').text)
        except NoSuchElementException:
            ret_exch.append('-')

        # Extracting expected delivery date     
        try:
            exp_del.append(driver.find_element_by_xpath("//div[@id='ddmDeliveryMessage']/b").text)
        except NoSuchElementException:
            exp_del.append('-')

        # Extracting Product availability data
        try:
            product_avail.append(driver.find_element_by_id('availability').text.replace('\n',''))
        except NoSuchElementException:
            product_avail.append('-')

        # Extracting Feature details of the product
        try:
            other_details.append(driver.find_element_by_xpath("//ul[@class='a-unordered-list a-vertical a-spacing-mini']").text.replace('\n','**--**'))
        except NoSuchElementException:
            other_details.append('-')
```

# Q7

```python
# creating a URL variable to store website URL
url_digit=('https://www.digit.in/')

# Establishing connection with the website
driver.get(url_digit)

# Finding the icon to open search bar and clicking it
driver.find_element_by_class_name('search').click()
time.sleep(1)

# Creating object of the search bar by finding it through Class name
srch_bar=driver.find_element_by_id('globalPageSearchText')

# Clearing the bar if there is any text in it
srch_bar.clear()

# Sending required product to the search bar
srch_bar.send_keys('top 10 gaming laptops')

# Sending Enter command to search for results
srch_bar.send_keys(Keys.ENTER)
time.sleep(1)

# Filtering the results for products only, thus finding and clicking products radio button
driver.find_element_by_xpath("//div[@class='whatSearch']/form/ul/li[2]").click()
time.sleep(2)

# Creating empty lists to store product data

prod_title=[]
prod_price=[]
prod_os=[]
prod_disp=[]
prod_processor=[]
prod_memory=[]
launch_status=[]
launch_date=[]
digit_rating=[]
prod_rating=[]
prod_desc=[]
prod_url=[]

# Fetching product URLs of the displayed results
for i in driver.find_elements_by_xpath("//div[@class='searchPage']/a"):
    prod_url.append(i.get_attribute('href'))
time.sleep(1)

# Iterating through the URLs to fetch product details
for link in prod_url:
    driver.get(link)
    time.sleep(2)
    
    #Extracting Product Title
    try:
        prod_title.append(driver.find_element_by_xpath("//div[@class='heading-wraper']/h1").text)
    except NoSuchElementException:
        prod_title.append('-')
        
    # Extracting Price
    try:
        prod_price.append(driver.find_element_by_xpath("//div[@class='Block-price']/b").text)
    except NoSuchElementException:
        prod_price.append('-')
    
    # Extracting Product Operating System
    try:
        prod_os.append(driver.find_element_by_xpath("//div[@class='Specs-Wrap  notspecs ']/ul/li[1]/div").text)
    except NoSuchElementException:
        prod_os.append('-')
    
    # Extracting Product Display
    try:
        prod_disp.append(driver.find_element_by_xpath("//div[@class='Specs-Wrap  notspecs ']/ul/li[2]/div").text)
    except NoSuchElementException:
        prod_disp.append('-')
    
    # Extracting Product Processor
    try:
        prod_processor.append(driver.find_element_by_xpath("//div[@class='Specs-Wrap  notspecs ']/ul/li[3]/div").text)
    except NoSuchElementException:
        prod_processor.append('-')
        
    # Extracting Product Memory
    try:
        prod_memory.append(driver.find_element_by_xpath("//div[@class='Specs-Wrap  notspecs ']/ul/li[4]/div").text)
    except NoSuchElementException:
        prod_memory.append('-')   
        
    # Extracting Launch Status
    try:
        launch_status.append(driver.find_element_by_xpath("//div[@class='status-bar only-desktop']/div[1]/b").text)
    except NoSuchElementException:
        launch_status.append('-')
        
    # Extracting launch date
    try:
        launch_date.append(driver.find_element_by_xpath("//div[@class='status-bar only-desktop']/div[2]/b").text)
    except NoSuchElementException:
        launch_date.append('-')
    
    #Extracting Digit Rating
    try:
        digit_rating.append(driver.find_element_by_xpath("//div[@class='Review-reting']/div[1]").text)
    except NoSuchElementException:
        digit_rating.append('-')
    
    # Extracting Ratings
    try:
        prod_rating.append(driver.find_element_by_xpath("//div[@class='Over-allrating']/div[2]").text.replace('/',' out of '))
    except NoSuchElementException:
        prod_rating.append('-')
        
    # Extracting Description
    try:
        prod_desc.append(driver.find_element_by_xpath("//div[@class='MoreLink-briefdescription']/div").text)
    except NoSuchElementException:
        prod_desc.append('-')

# Creating Dataframe
prod_digit=pd.DataFrame(list(zip(prod_title,prod_price,prod_os,prod_disp,prod_processor,prod_memory,launch_status,launch_date,digit_rating,prod_rating,prod_desc,prod_url)),columns=['Product_Title','Product_Price','Product_OS','Product_Display','Product_Processor','Product_Memory','Launch_Status','Launch_Date','Digit_Rating','User_Rating','Product Description','Product_Url'])
prod_digit

```

```python
# Exporting to a CSV file
prod_digit.to_csv('gaming_laptops_digit.csv')
```
