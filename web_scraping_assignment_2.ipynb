#  Q1. Scraping Data Analyst Job Data from Naukri.com

```python
# Creating empty lists to store data

job_title=[]
job_location=[]
company_name=[]
experience_req=[]

# creating a URL variable to store website URL

url=('https://www.naukri.com/')

# Defining a function to scrape data with the required job title and location

def job_search(title,location):
    
    # Establishing Connection with the website
    driver.get(url)
    time.sleep(5) # pausing the code execution for 5 seconds for website to load
    
    # Creating object of the search bar by finding it through ID
    search_bar=driver.find_element_by_id('qsb-keyword-sugg')
    
    # Clearing the bar if there is any text in it
    search_bar.clear()
    
    # Sending required designation to the search bar
    search_bar.send_keys(title)
    
    # Creating object for the location bar by finding it through ID
    location_bar=driver.find_element_by_id('qsb-location-sugg')
    
    # Clearing the bar if there is any text in it
    location_bar.clear()
    
    # Sending required location to the search bar
    location_bar.send_keys(location)
    
    # Creating onject of the search button and sending click commamnd
    search_btn=driver.find_element_by_xpath("//button[@class='btn']")
    search_btn.click()
    time.sleep(4) # pausing the code execution for website to load
    
    # Creating for Loop to extract Job Title
    for title in driver.find_elements_by_xpath("//a[@class='title fw500 ellipsis']")[0:10]:
        job_title.append(title.text)
    
    # Creating for loop to extract company name
    for company in driver.find_elements_by_xpath("//a[@class='subTitle ellipsis fleft']")[0:10]:
        company_name.append(company.text)
    
    # Creating for loop to extract job location
    for location in driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi location']/span")[0:10]:
        job_location.append(location.text)
    
    # Creating for loop to extract required experience
    for exp in driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi experience']/span")[0:10]:
        experience_req.append(exp.text)
    
```

```python
# Calling the function with required inputs
job_search('data analyst','bangalore')
```

```python
# Creating Dataframe
data_analyst=pd.DataFrame(list(zip(job_title,job_location,company_name,experience_req)),columns=['Job_Title','Company_name','Experience_Required','Job_Location'])
```

# Q2. Scraping Data Scientist Job Data from Naukri.com

```python
# Creating empty lists to store data

job_title_s=[]
job_location_s=[]
company_name_s=[]
job_desc_s=[]

# creating a URL variable to store website URL

url2=('https://www.naukri.com/')

# Defining a function to scrape data with the required job title and location

def find_job(title,location):
    
    # Establishing Connection with the website
    driver.get(url2)
    time.sleep(5) # pausing the code execution for 5 seconds for website to load
    
    # Creating object of the search bar by finding it through ID
    search_bar=driver.find_element_by_id('qsb-keyword-sugg')
    
    # Clearing the bar if there is any text in it
    search_bar.clear()
    
    # Sending required designation to the search bar
    search_bar.send_keys(title)
    
    # Creating object for the location bar by finding it through ID
    location_bar=driver.find_element_by_id('qsb-location-sugg')
    
    # Clearing the bar if there is any text in it
    location_bar.clear()
    
    # Sending required location to the search ba
    location_bar.send_keys(location)
    
    # Creating object of the search button and sending click commamnd
    search_btn=driver.find_element_by_xpath("//button[@class='btn']")
    search_btn.click()
    time.sleep(4)
    
    # Creating a variable to scrape and store URLs of Job Titles
    url1=[]
    
    # Creating a for loop to extract URLs
    for link in driver.find_elements_by_xpath("//a[@class='title fw500 ellipsis']")[0:10]:
        url1.append(link.get_attribute('href'))
    
    # Creating a for loop to iterate through URL list and visit each URL to fetch data
    for url in url1[0:10]:
        driver.get(url)
        time.sleep(5)
        
        # Extracting Job title
        job_title_s.append(driver.find_element_by_xpath("//h1[@class='jd-header-title']|//h1[@class='av-special-heading-tag ']").text)
        
        # Extracting Company Name
        company_name_s.append(driver.find_element_by_xpath("//a[@class='pad-rt-8']|//p[@class='cpName f14']").text)
        
        # Extracting Job Location
        job_location_s.append(driver.find_element_by_xpath("//span[@class='location ']|//span[@class='slide-meta-loc pull-left']").text)
        
        # Extracting Description of the Job
        job_desc_s.append(driver.find_element_by_xpath("//section[@class='job-desc']|//section[@class='JD av_textblock_section']").text.replace('\n','*-*-*'))
```

```python
# Calling function with required inputs

find_job('data scientist','bangalore')
```

```python
# Creating a dataframe to store data
data_scientist=pd.DataFrame(list(zip(job_title_s,company_name_s,job_location_s,job_desc_s)),columns=['Job_Title','Company_Name','Job_Location','Job_Description'])
```

# Defining Function to scrap data from flipkart

```python

# creating a URL variable to store website URL
url_fk=('https://www.flipkart.com/')

# Defining a function to scrap details about any product with no of pages to be scraped
def product_fk(product,no_page):
    
    # Establishing Connection with the website
    driver.get(url_fk)
    time.sleep(5) # pausing the code execution for 5 seconds for website to load
    
    # Closing the login dialogue box appearing at first visit
    try:
        driver.find_element_by_xpath("//button[@class='_2KpZ6l _2doB4z']").click()
    except NoSuchElementException:
        pass

    # Creating object of the search bar by finding it through Class name
    search_bar_fk=driver.find_element_by_class_name('_3704LK')

    # Clearing the bar if there is any text in it
    search_bar_fk.clear()

    # Sending required product to the search bar
    search_bar_fk.send_keys(product)

    # Creating object of the search button and sending click commamnd
    search_btn_fk=driver.find_element_by_class_name('_34RNph')
    search_btn_fk.click()
    time.sleep(3) # pausing the code for 3 seconds, allowing page to load

    # Creating a for loop to iterate through pages and fetch required data

    for page in range(0,no_page):
    
        # Creating for Loop to extract Brand name    
        for brand in driver.find_elements_by_xpath("//div[@class='_2WkVRV']"):
            brand_name_fk.append(brand.text)
    
        # Creating for Loop to extract Product Description    
        for desc in driver.find_elements_by_xpath("//a[@class='IRpwTa']"):
            desc_fk.append(desc.text)
    
        # Creating for Loop to extract Product Price
        for price in driver.find_elements_by_xpath("//div[@class='_30jeq3']")[0:40]:
            price_fk.append(price.text)
    
        # Creating for Loop to extract Discount on Product
        for disc in driver.find_elements_by_xpath("//div[@class='_3Ay6Sb']")[0:40]:
            discount_fk.append(disc.text)
    
        # Creating for Loop to iterate to next page    
        for i in driver.find_elements_by_xpath("//a[@class='_1LKTO3']"):
            try:
                driver.find_element_by_xpath("//a[@class='_1LKTO3'][2]").click()
                time.sleep(5)
            except NoSuchElementException:
                driver.find_element_by_xpath("//a[@class='_1LKTO3'][1]").click()
                time.sleep(5)
```

# Q9. Scraping Data from Myntra

```python
# Creating empty lists to store data
product_brand_myn=[]
product_desc_myn=[]
product_price_myn=[]

# creating a URL variable to store website URL

url_myn=('https://www.myntra.com/shoes')

# Establishing Connection with the website
driver.get(url_myn)
time.sleep(3)

# Creating For loop to find and select required checkboxes
for color in driver.find_elements_by_xpath("//label[@class='common-customCheckbox']"):
    if 'Black' in color.text:
        color.click()
time.sleep(2)

for price in driver.find_elements_by_xpath("//label[@class='common-customCheckbox vertical-filters-label']"):
    if 'Rs. 6649 to Rs. 13099' in price.text:
        price.click()
time.sleep(2)

# Creating For loop to Scrape data and iterate through pages
for i in range(0,2):
    
    # Creating for Loop to extract Brand Name
    for brand in driver.find_elements_by_xpath("//h3[@class='product-brand']"):
        product_brand_myn.append(brand.text)
    
    # Creating for Loop to extract Product Description
    for desc in driver.find_elements_by_xpath("//h4[@class='product-product']"):
        product_desc_myn.append(desc.text)
        
    # Creating for Loop to extract Product Price
    
    products=driver.find_elements_by_xpath("//div[@class='product-price']") # storing all elements of parent DIV tag
    
    # Creating a loop to iterate though the list
    for product in products:
        # From the elements stored in products, fetching elements with discounted price
        discount_price=product.find_elements_by_xpath(".//span[@class='product-discountedPrice']") 
        # Defining a condition to match in discount_price. For Boolean True, If statement will work and for False, else will work.
        if (discount_price):
            product_price_myn.append(discount_price[0].text)
        else:
            product_price_myn.append(product.find_element_by_xpath(".//span").text)
    
    # Finding the element for Next button and clicking it
    driver.find_element_by_xpath("//li[@class='pagination-next']/a").click()
    time.sleep(3)
```

```python
# Creating Dataframe

shoes_myn=pd.DataFrame(list(zip(product_brand_myn,product_desc_myn,product_price_myn)),columns=['Brad_Name','Product_Description','Product_Price'])
shoes_myn
```

```python
# Exporting to a CSV file
shoes_myn.to_csv('shoes_myn.csv')
```

# Q.10 Scraping Laptop Data from Amazon

```python
# Creating empty lists to store data
product_title=[]
product_rating=[]
product_price=[]

# creating a URL variable to store website URL
url_amzn=('https://www.amazon.in/')

# Establishing Connection with the website
driver.get(url_amzn)
time.sleep(2) # pausing the code execution for 2 seconds for website to load

# Creating object of the search bar by finding it through ID
srch_bar=driver.find_element_by_id('twotabsearchtextbox')

# Clearing the bar if there is any text in it
srch_bar.clear()

# Sending required product to the search bar
srch_bar.send_keys('laptops')

# Creating onject of the search button and sending click commamnd
srch_btn=driver.find_element_by_id('nav-search-submit-button')
srch_btn.click()
time.sleep(2)

# Creating a For loop to get checkboxes and click the one with I7 as text
for cpu in driver.find_elements_by_xpath("//a[@class='a-link-normal s-navigation-item']"):
    if 'i7' in cpu.text:
        cpu.click()
        break
time.sleep(2)

# Collecting URLs for each product on the page
links=[]
for i in driver.find_elements_by_xpath("//h2[@class='a-size-mini a-spacing-none a-color-base s-line-clamp-2']/a")[0:10]:
    links.append(i.get_attribute('href'))

# Accessing each product through URL
for link in links:
    driver.get(link)
    time.sleep(2)
    
    # Extracting Product Title
    product_title.append(driver.find_element_by_xpath("//span[@class='a-size-large product-title-word-break']").text)
    
    # Extracting Product Price. There are three different prices DealPrice, Amazon'sPrice, SalePrice. We search for anyone of the following. If none is found we append No Price available.
    try:
        product_price.append(driver.find_element_by_xpath("//span[@class='a-size-medium a-color-price priceBlockDealPriceString']|//span[@class='a-size-medium a-color-price priceBlockBuyingPriceString']|//span[@class='a-size-medium a-color-price priceBlockSalePriceString']").text)
        
    except NoSuchElementException:        
        product_price.append('Price Not Available')
    
    # Extracting Product Rating
    try:
        product_rating.append(driver.find_element_by_xpath("//span[@class='a-size-base a-nowrap']/span").text)
        
    except NoSuchElementException:
        product_rating.append('-')

# Getting to laptops result page again
driver.get('https://www.amazon.in/s?k=laptops&ref=nb_sb_noss')
time.sleep(2)

# Creating a For loop to get checkboxes and click the one with I9 as text
for cpu in driver.find_elements_by_xpath("//a[@class='a-link-normal s-navigation-item']"):
    if 'i9' in cpu.text:
        cpu.click()
        break
time.sleep(2)

# Collecting URLs for each product
links=[]
for i in driver.find_elements_by_xpath("//h2[@class='a-size-mini a-spacing-none a-color-base s-line-clamp-2']/a")[0:10]:
    links.append(i.get_attribute('href'))

# Accessing each product through URL
for link in links:
    driver.get(link)
    time.sleep(2)
    
    # Extracting Product Title
    product_title.append(driver.find_element_by_xpath("//span[@class='a-size-large product-title-word-break']").text)
    
    # Extracting Product Price. There are three different prices DealPrice, Amazon'sPrice, SalePrice. We search for anyone of the following. If none is found we append No Price available.
    try:
        product_price.append(driver.find_element_by_xpath("//span[@class='a-size-medium a-color-price priceBlockDealPriceString']|//span[@class='a-size-medium a-color-price priceBlockBuyingPriceString']|//span[@class='a-size-medium a-color-price priceBlockSalePriceString']").text)
        
    except NoSuchElementException:        
        product_price.append('Price Not Available')
    
    # Extracting Product Rating
    try:
        product_rating.append(driver.find_element_by_xpath("//span[@class='a-size-base a-nowrap']/span").text)
    except NoSuchElementException:
        product_rating.append('-')
```

```python
# Creating Dataframe
laptop_amzn=pd.DataFrame(list(zip(product_title,product_rating,product_price)),columns=['Product_Name','Product_Price','Product_Rating'])
laptop_amzn
```

```python
# Exporting to a CSV file
laptop_amzn.to_csv('laptop_amzn.csv')
```
