# Q1

```python
# Creating a URL variable to store website URL
url_wiki=('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')

# Establishing Connection with the website
driver.get(url_wiki)
time.sleep(2)

# Creating empty lists to store data

video_rank=[]
video_name=[]
video_uploader=[]
upload_date=[]
video_views=[]

# Finding and storing table rows which contains data in a variable
data1=driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr")

# Creating a For loop to iterate through the rows to find required data
for data in data1:
    # Finding Rank
    video_rank.append(data.find_element_by_xpath("td[1]").text.replace('.',''))
    
    # Finding Video Title
    video_name.append(data.find_element_by_xpath('td[2]').text.split('\"')[1])
    
    # Finding Uploader's name
    video_uploader.append(data.find_element_by_xpath("td[3]").text)
    
    # Finding Upload Date
    upload_date.append(data.find_element_by_xpath("td[5]").text)
    
    # Finding views on the video
    video_views.append(data.find_element_by_xpath("td[4]").text)

# Creating Dataframe

yt_views=pd.DataFrame(list(zip(video_rank,video_name,video_uploader,upload_date,video_views)),columns=['Rank','Video_Title','Video_Uploader','Upload_Date','Views (In Billions)'])
yt_views
```

```python
# Exporting to a CSV file
yt_views.to_csv("yt_views.csv")
```

```python
# Creating a URL variable to store website URL
url_bcci=("https://www.bcci.tv/")

# Establishing Connection with the website
driver.get(url_bcci)
time.sleep(3)

# Creating empty lists to store data
match_title=[]
series=[]
match_type=[]
match_location=[]
match_date=[]
match_time=[]

# Finding the International dropdown and hover over it
drop_down=driver.find_element_by_xpath("//div[@class='navigation__link navigation__link--has-drop-down js-navigation-link']")
time.sleep(1)
action.move_to_element(drop_down).perform()
time.sleep(1)

# Clicking the fixtures options
driver.find_element_by_xpath("//div[@class='drop-down__options']/ul/li[1]").click()

# Extracting Match Title
for title in driver.find_elements_by_xpath("//p[@class='fixture__additional-info']/strong"):
    match_title.append(title.text)

# Extracting Series Name
for ser in driver.find_elements_by_xpath("//div[@class='fixture__format-strip']/span[2]"):
    series.append(ser.text)

# Extracting Match Type 
for form in driver.find_elements_by_xpath("//div[@class='fixture__format-strip']/span[1]"):
    match_type.append(form.text)

# Extracting Match Location
for loc in driver.find_elements_by_xpath("//p[@class='fixture__additional-info']/span"):
    match_location.append(loc.text)

#Extracting Match Date
for dt in driver.find_elements_by_xpath("//div[@class='fixture__datetime desktop-only']"):
    match_date.append(dt.text.rsplit(maxsplit=2)[0].replace('\n',' '))

# Extracting Match Time
for time1 in driver.find_elements_by_xpath("//span[@class='fixture__time']"):
    match_time.append(time1.text)

# Creating Dataframe
bcci_fixtures=pd.DataFrame(list(zip(match_title,series,match_type,match_location,match_date,match_time)),columns=['Title','Series','Format','Location','Date','Time'])
bcci_fixtures
```

```python
# creating a URL variable to store website URL
url_ics=('https://archive.ics.uci.edu/')

# Establishing Connection with the website
driver.get(url_ics)
time.sleep(3)

# Finding and clicking View All Dataset option 
driver.find_element_by_xpath("//span[@class='whitetext'][2]/a").click()

# Creating empty list
dataset_name=[]
dataset_type=[]
dataset_task=[]
dataset_attrib_types=[]
dataset_instances=[]
dataset_atrribute=[]
dataset_year=[]

# Creating a For loop to iterate through the tags and extract required data
for repo in driver.find_elements_by_xpath("//body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr")[1:]:
    
    # Extracting Dataset name
    dataset_name.append(repo.find_element_by_xpath("td[1]/table/tbody/tr/td[2]/p/b/a").text)
    
    # Extracting Data type
    ds_type=repo.find_element_by_xpath("td[2]/p").text
    if ds_type != ' ':
        dataset_type.append(ds_type)
    else:
        dataset_type.append('-')
    
    # Extracting Dataset task
    ds_task=repo.find_element_by_xpath("td[3]/p").text
    if ds_task != ' ':
        dataset_task.append(ds_task)
    else:
        dataset_task.append('-')
    
    # Extracting Dataset atrribute types
    ds_attrib_type=repo.find_element_by_xpath("td[4]/p").text
    if ds_attrib_type != ' ':
        dataset_attrib_types.append(ds_attrib_type)
    else:
        dataset_attrib_types.append('-')
    
    # Extracting Dataset instances   
    ds_inst=repo.find_element_by_xpath("td[5]/p").text
    if ds_inst != ' ':
        dataset_instances.append(ds_inst)
    else:
        dataset_instances.append('-')
    
    # Extracting Dataset Attributes
    ds_attrib=repo.find_element_by_xpath("td[6]/p").text
    if ds_attrib != ' ':
        dataset_atrribute.append(ds_attrib)
    else:
        dataset_atrribute.append('-')
    
    # Extracting Year    
    ds_year=repo.find_element_by_xpath("td[7]/p").text
    if ds_year != ' ':
        dataset_year.append(ds_year)
    else:
        dataset_year.append('-')

# Creating Dataframe
datasets_uci=pd.DataFrame(list(zip(dataset_name,dataset_type,dataset_task,dataset_attrib_types,dataset_instances,dataset_atrribute,dataset_year)),columns=['Name','Data_Types','Task','Atrribute_Types','Instances','Atrributes','Year'])
datasets_uci
```
