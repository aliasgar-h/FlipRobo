# IMDB Gobal Top 100 Movies

```python
# Defining empty lists for storing Movie titles, Serial No, Year of Relaese and Rating respectively

movie_name=[]
serial_no=[]
yor=[]
movie_rating=[]

# Defining a  function to extract data from both pages
    
def imdb_global(url_1,url_2):
    
    # Importing content of the webpages
    
    data1=requests.get(url_1)
    data2=requests.get(url_2)
    
    # Creating BeautifulSoup objects
    
    imdb_global_1=BeautifulSoup(data1.content)
    imdb_global_2=BeautifulSoup(data2.content)
    
    
    # Extracting all the H3 Header tags which contains serial no, Movie Title and Year of release
        
    movie_title_1=imdb_global_1.find_all('h3',class_="lister-item-header")
    movie_title_2=imdb_global_2.find_all('h3',class_="lister-item-header")
        
    # Extracting movie title
        
    for m1 in movie_title_1:
        movie_name.append(m1.find('a').text)
        
    for m2 in movie_title_2:
        movie_name.append(m2.find('a').text)
    
    # Extracting Serial No
    
    for s1 in movie_title_1:
        serial_no.append(s1.find('span',class_='lister-item-index').text.replace('\n',''))
    
    for s2 in movie_title_2:
        serial_no.append(s2.find('span',class_='lister-item-index').text.replace('\n',''))
    
    # Extracting Year of Release
    
    for y1 in movie_title_1:
        yor.append(y1.find('span',class_='lister-item-year').text)
    
    for y2 in movie_title_2:
        yor.append(y2.find('span',class_='lister-item-year').text.replace('(I) ',''))
    
    # Extracting Div tag which contains Rating of the movie
    
    rating_1=imdb_global_1.find_all('div',class_='ratings-imdb-rating')
    rating_2=imdb_global_2.find_all('div',class_='ratings-imdb-rating')
    
    # Extracting Rating
    
    for r1 in rating_1:
        movie_rating.append(r1.find('strong').text) 
    
    for r2 in rating_2:
        movie_rating.append(r2.find('strong').text)
    
```

```python
# Calling the function with URLs of both pages

imdb_global('https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc','https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc&start=51&ref_=adv_nxt')
```

```python
# Creating Dataframe

imdb_global_top_100=pd.DataFrame(list(zip(serial_no,movie_name,yor,movie_rating)),columns=['S.no','Movie_Title','Year_of_Release','Imdb_Rating'])

imdb_global_top_100['Year_of_Release']=imdb_global_top_100['Year_of_Release'].str.strip("[()]")
```

```python
imdb_global_top_100
```

```python
# Exporting to csv file

imdb_global_top_100.to_csv('imdb_global_top_100.csv')
```

# IMDB Indian Top 100 Movies

```python
# Defining empty lists for storing Movie titles, Serial No, Year of Relaese and Rating respectively

movie_title=[]
s_no=[]
y_o_r=[]
imdb_rating=[]

# Defining a  function to extract data
    
def imdb_indian(url):
    
    # Importing content of the webpages
    
    data=requests.get(url)
        
    # Creating BeautifulSoup object
    
    imdb_indian=BeautifulSoup(data.content)
       
    
    # Extracting all the H3 Header tags which contains serial no, Movie Title and Year of release
        
    title=imdb_indian.find_all('h3',class_="lister-item-header")
    
    
    # Extracting movie title
        
    for m in title:
        movie_title.append(m.find('a').text)
        
        
    # Extracting Serial No
    
    for s in title:
        s_no.append(s.find('span',class_='lister-item-index').text.replace('.',''))

    
    # Extracting Year of Release
    
    for y in title:
        y_o_r.append(y.find('span',class_='lister-item-year').text.replace('(I) ',''))
    
    
    # Extracting Div tag which contains Rating of the movie
    
    rating=imdb_indian.find_all('div',class_='ipl-rating-widget')
        
    # Extracting Rating
    
    for r in rating:
        imdb_rating.append(r.find('span',class_='ipl-rating-star__rating').text)
    
```

```python
# Calling the function with URL of the page

imdb_indian('https://www.imdb.com/list/ls009997493/?sort=user_rating,desc&st_dt=&mode=detail&page=1')
```

```python
# Cresting Dataframe

imdb_indian_top_100=pd.DataFrame(list(zip(s_no,movie_title,y_o_r,imdb_rating)),columns=['S.no','Movie_Title','Year_of_Release','Imdb_Rating'])

imdb_indian_top_100['Year_of_Release']=imdb_indian_top_100['Year_of_Release'].str.strip("[()]")
```

```python
# Exporting to csv file

imdb_indian_top_100.to_csv('imdb_indian_top_100.csv')
```

# Book List

```python
# Importing webpage

bookpage=requests.get('https://bookpage.com/reviews')
bookpage
```

```python
# Creating BeautifulSoup object

books=BeautifulSoup(bookpage.content)
```

```python
# Cretaing List to store book title

book_title=[]
author_name=[]
book_genre=[]
book_review=[]

# Extracting tags which contains book title

title=books.find_all('h4',class_='italic')

# Extracting Title of Books

for t in title[:5]:    
    book_title.append(t.find('a').text.replace('  ',''))

# Extracting Tags which contains Author name

auth_name=books.find_all('p',class_='sans bold')

# Extracting Author Name

for a in auth_name[:5]:
    author_name.append(a.text.replace('\n',''))

# Extracting Tags which contains Genre data

genre=books.find_all('p',class_='genre-links')

# Extracting Genre of Books

for g in genre[:5]:
    book_genre.append(g.text.replace('\n',''))

# Extracting Tags which contains Review data

review=books.find_all('p',class_='excerpt')

# Extracting Book Reviews

for r in review[:5]:
    book_review.append(r.text.replace('\n',''))
```

```python
# Creating Dataframe

book_page=pd.DataFrame(list(zip(book_title,author_name,book_genre,book_review,)),columns=['Book Title','Author','Genre','Review'])
```

```python
book_page
```

```python
# Exporting to CSV file

book_page.to_csv('bookpage.csv')
```

# ICC Cricket Data

ICC Men's Team Ranking

```python
# Importing Webpage

icc_odi_m_team=requests.get('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')
icc_odi_m_team
```

```python
# Defining lists to store Position, Team Name, Number of Matches, Points and Icc Rating for Teams in ODI

team_position=[]
team_name=[]
team_matches=[]
team_points=[]
team_rating=[]

# Defining Funtion to extract ICC Team Data

def icc_odi_team(url_variable):
    
    # Creating BeautifulSoup Object
    
    icc_men_team=BeautifulSoup(url_variable.content)
    
    # Extracting tags which contains required data
    
    data=icc_men_team.find_all('tr',class_='rankings-block__banner')
    data2=icc_men_team.find_all('tr',class_='table-body')
    data3=icc_men_team.find_all('td',class_='u-center-text')
        
    
    # Extracting Team Position
        
    for p in data:
        team_position.append(p.find('td',class_='rankings-block__banner--pos').text)
    
    for p2 in data2[:9]:
        team_position.append(p2.find('td',class_='u-text-right').text)
        
    
    # Extracting Team Name
    
    for n in data:
        team_name.append(n.find('span',class_='u-hide-phablet').text)
        
    for n2 in data2[:9]:
        team_name.append(n2.find('span',class_='u-hide-phablet').text)
        
    
    # Extracting Number of Matches and Points
    
    # Extracting Matches for top team
    
    for m in data:
        team_matches.append(m.find('td',class_='rankings-block__banner--matches').text)
    
    # Extracting Points of Top Team
    
    for o in data:
        team_points.append(o.find('td',class_='rankings-block__banner--points').text)
    
    # Extracting matches and points for rest of the teams
    
    for count in range(0,len(data3[:18])):
        if count % 2:
            team_points.append(data3[count].text)
        else:
            team_matches.append(data3[count].text)
       
    
    # Extracting Rating
    
    for r in data:
        team_rating.append(r.find('td',class_='rankings-block__banner--rating').text.replace('\n','').strip())
        
    for r2 in data2[:9]:
        team_rating.append(r2.find('td',class_='rating').text)
    
```

```python
# Calling Function with url variable

icc_odi_team(icc_odi_m_team)
```

```python
# Creating Dataframe

odi_team_m=pd.DataFrame(list(zip(team_position,team_name,team_matches,team_points,team_rating)),columns=['Pos','Team','Matches','Points','Rating'])
```

```python
odi_team_m
```

```python
# Exporting to a csv file

odi_team_m.to_csv('odi_team_m.csv')
```

ICC Men's Batsmen ODI ranking

```python
# Importing Webpage

batsmen=requests.get('https://www.icc-cricket.com/rankings/mens/player-rankings/odi/batting')
batsmen
```

```python
# Defining lists to store Player Position, Player Name, Nationality and Icc Rating for Batsmen in ODI

batsmen_position=[]
batsmen_name=[]
batsmen_nat=[]
batsmen_rating=[]

# Defining Funtion to extract ICC Batsmen Data

def icc_odi_batsmen(url_variable):
    
    # Creating BeautifulSoup Object
    
    batsmen_ranking=BeautifulSoup(url_variable.content)
    
    # Extracting Tags which contains required data
    
    data=batsmen_ranking.find_all('tr',class_='rankings-block__banner')
    data1=batsmen_ranking.find_all('tr',class_='table-body')
    
    # Extracting Player Position
    
    for p in data:
        batsmen_position.append(p.find('td',class_='rankings-block__position').text)
        
    for p1 in data1:
        batsmen_position.append(p1.find('td',class_='u-text-right').text.replace('\n','').replace('This player has moved down in the rankings since the previous rankings update','').replace('This player has moved up in the rankings since the previous rankings update','').strip())
    
    # Extracting Player name
    
    for n in data:
        batsmen_name.append(n.find('div',class_='rankings-block__banner--name-large').text)
    
    for n1 in data1[:9]:
        batsmen_name.append(n1.find('td',class_='name').text.replace('\n',''))
    
    # Extracting Player's Nationality
    
    for c in data:
        batsmen_nat.append(c.find('div',class_='rankings-block__banner--nationality').text.replace('\n','').strip())
    
    for c1 in data1[:9]:
        batsmen_nat.append(c1.find('span',class_='table-body__logo-text').text)
        
    # Extracting Rating
    
    for r in data:
        batsmen_rating.append(r.find('div',class_='rankings-block__banner--rating').text)
        
    for r1 in data1[:9]:
        batsmen_rating.append(r1.find('td',class_='rating').text)
        
```

```python
# Calling Funtion with url variable

icc_odi_batsmen(batsmen)
```

```python
# Creating Dataframe

odi_batsmen_m=pd.DataFrame(list(zip(batsmen_position,batsmen_name,batsmen_nat,batsmen_rating)),columns=['Position','Player','Team','Rating'])
```

```python
# Defining lists to store Mobile Model, Mobile Price, Image Url and Average Rating for Mobiles

mobile_model=[]
mobile_price=[]
mobile_image=[]
mobile_rating=[]

# Defining Function to Extract data

def mobile_amazon_20k(url_variable):
    
    # Creating BeautifulSoup object
    
    mobile=BeautifulSoup(url_variable.content)
    
    # Extracting Mobile model name
    
    #Extracting tags which contains model name
    
    model=mobile.find_all('span',class_='a-size-medium')
    
    # Extracting model names
    
    for m in model:
        mobile_model.append(m.text)
    
    
    # Extracting Mobile Price
    
    #Extracting tags which contains Mobile prices
    
    price=mobile.find_all('span',class_='a-price-whole')    
    
    # Extracting Mobile Price
    
    for p in price:
        mobile_price.append(p.text)
    
    
    # Extracting Image Urls
    
    #Extracting tags which contains Mobile Image Urls
    
    image=mobile.find_all('img',class_='s-image')    
    
    # Extracting Urls of the Mobile Image
    
    for i in image:
        mobile_image.append(i.get('src'))
    
    
    # Extracting Average Mobile rating
    
    #Extracting tags which Average Ratings
    
    rating=mobile.find_all('span',class_='a-icon-alt')    
    
    # Extracting Average Ratings
    
    for r in rating:
        mobile_rating.append(r.text)
    
```

```python
# Calling function with Url

mobile_amazon_20k(mob)
```

# Weather Forecast Data

```python
# Importing Webpage

weather=requests.get('https://forecast.weather.gov/MapClick.php?lat=37.777120000000025&lon=-122.41963999999996')
weather
```

```python
# Defining lists to store Period, Weather_Condition, Temperature and Detailed weather condition

period=[]
short_desc=[]
temp=[]
desc=[]

# Defining funtion to extract Temp details from USA Weather Forecast Website

def weather_usa(url_variable):
    
    # Creating a BeautifulSoup instance
    
    data=BeautifulSoup(url_variable.content)
    
    # Extracting tags which contains required data
    
    data1=data.find_all('div',class_='tombstone-container')
    data2=data.find_all('div',class_='forecast-text')
    
    # Extracting Period Data
    
    for p in data1:
        period.append(p.find('p',class_='period-name').text)
    
    # Extracting Weather condition
    
    for sh in data1:
        short_desc.append(sh.find('p',class_='short-desc').text)
    
    # Extracting Temperature Readings
    
    for tmp in data1:
        temp.append(tmp.find('p',class_='temp').text)
    
    # Extracting Description about the weather condition
    
    for des in data2:
        desc.append(des.text)
```

```python
# Calling Function with url variable

weather_usa(weather)
```

```python
# Creating Dataframe

weather_usa=pd.DataFrame(list(zip(period,short_desc,temp,desc)),columns=['Period','Condition','Temperature','Details'])
```

```python
weather_usa
```

```python
# Exporting to CSV file

weather_usa.to_csv('weather_usa.csv')
```
